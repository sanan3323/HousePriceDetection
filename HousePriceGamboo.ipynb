```python
total = df.isnull().sum().sort_values(ascending=False)
percent = (df.isnull().sum()/df.isnull().count()).sort_values(ascending=False)
missing_data = pd.concat([total, percent], axis=1, keys=['Total', 'Percent'])
missing_data.head(20)
```

```python
df = df.drop((missing_data[missing_data['Total'] > 80]).index,1)
```

```python
df1 = df.fillna(df.mean())
```

```python
num_col = [col for col in df1.columns if df1[col].nunique()>10 and df1[col].dtype in ['int64','float64']]
num_col
```

```python
df1['Street'].replace({'Pave':0,
                       'Grvl':1},inplace=True)
df1['Utilities'].replace({'AllPub':0,
                          'NoSeWa':1},inplace=True)
```

```python
cat_cols = ['Zoningof MS', 'LotOFShape', 'LandContour', 'LotOFConfig', 'SlopeOFland','Neighbor', 'firstCondition', 'secondCondition', 
            'TypeofBLDG', 'StyleOFhouse', 'Styleofroof', 'RoofofMatl','Exteriorfirst', 'Exteriorsecond', 'MasVnrType',
            'ExterQualifaction' ,'ExterCondition','Foundation','BsmtQual','BsmtCond',
            'BsmtExposure','BsmtFinType1', 'BsmtFinType2', 'Heating', 'HeatingQC', 'CentralAir','Electrical', 'KitchenQual', 
            'Functional', 'PavedDrive', 'SaleType', 'SaleCondition']
le = LabelEncoder()
for col in cat_cols:
    df1[col] = le.fit_transform(df1[col].astype(str))
```

```python
print(df1['Zoningof MS'].unique())
```

```python
df1.drop('Id',axis=1,inplace=True)
```

```python
x = df1.drop('SalePrice',axis=1)
y = df1['SalePrice']
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(x,y,shuffle=True,test_size=0.2)
```

```python
sc = StandardScaler()
xtrain = sc.fit_transform(xtrain)
xtest = sc.transform(xtest)
```

```python
def score(model,x_train,x_test,y_train,y_test):
  
  train_predict=model.predict(x_train)
  test_predict = model.predict(x_test)

  r2_train = r2_score(y_train,train_predict)
  r2_test = r2_score(y_test,test_predict)  #high r2score is target

  mse_train = mean_squared_error(y_train,train_predict)
  mse_test = mean_squared_error(y_test,test_predict)

  return (r2_train,r2_test,mse_train,mse_test)
```

```python
lr = LinearRegression()
lr.fit(xtrain,ytrain)
result1 = score(model=lr,x_train=xtrain,x_test=xtest,y_train=ytrain,y_test=ytest)
```

```python
print(f'Train R2 = {result1[0]}, Train MSE = {result1[2]} ')
print(f'Test R2 = {result1[1]}, Test MSE = {result1[3]} ')
```

```python
def evaluation(model_str, y_pred, y_pred_train):
    results = {
        'model': model_str,
        'R2 test' : r2_score(ytest, y_pred),
        'R2 train': r2_score(ytrain, y_pred_train),
        'MSE test' : mean_squared_error(ytest, y_pred),
        'MSE train' : mean_squared_error(ytrain, y_pred_train),
        'MAE test' : mean_absolute_error(ytest, y_pred),
        'MAE train' : mean_absolute_error(ytrain, y_pred_train)
    }
    return results
```

```python
model_dict = {
    'Linear Regression' : LinearRegression(),
    'Ridge': Ridge(),
    'Lasso': Lasso(),
    'ElasticNet': ElasticNet(),
    'RandomForestRegressor':RandomForestRegressor(),
    'DecisionTreeRegressor':DecisionTreeRegressor()
}
```

```python
%%time
result_list = []
for model in model_dict:
    model_dict[model].fit(xtrain,ytrain)
    y_pred = model_dict[model].predict(xtest)
    y_pred_train = model_dict[model].predict(xtrain)
    result = evaluation(model, y_pred, y_pred_train)
    result_list.append(result)
df_eval = pd.DataFrame(result_list)
df_eval
```

```python
rf = RandomForestRegressor()
param_grid = {
    'n_estimators': [100, 200, 300],
    'max_depth': [None, 5, 10],
    'min_samples_split': [2, 4, 8]
}


# Define the search method
grid_search = GridSearchCV(rf, param_grid, cv=5,scoring='neg_mean_squared_error')

# Fit the model with the best hyperparameters
grid_search.fit(xtrain, ytrain)

# Evaluate the performance of the model on the testing set
y_pred = grid_search.predict(xtest)
print('R2:', r2_score(ytest, y_pred))
print('MSE:', mean_squared_error(ytest, y_pred))
print('MAE:', mean_absolute_error(ytest, y_pred))
```
